```python
#Load libraries
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
from ctgan import CTGANSynthesizer
```

```python
HOME_PATH = '' #home path of the project
TRAIN_FILE = 'REAL DATASETS/TRAIN DATASETS/D_ContraceptiveMethod_Real_Train.csv'
SYNTHETIC_FILE = 'SYNTHETIC DATASETS/CTGAN/D_ContraceptiveMethod_Synthetic_CTGAN.csv'
```

## 1. Load the dataset

```python
real_data = pd.read_csv(HOME_PATH + TRAIN_FILE)
real_data
```

```python
real_data.dtypes
```

```python
discrete_columns = ['wife_education','husband_education','wife_religion','wife_working','husband_occupation',
                    'standard_of_living_index','media_exposure','contraceptive_method_used']
discrete_columns
```

```python
ctgan = CTGANSynthesizer()
ctgan.fit(real_data, discrete_columns)
```

```python
synthetic_data = ctgan.sample(len(real_data))
synthetic_data.to_csv(HOME_PATH + SYNTHETIC_FILE, index = False)
synthetic_data
```
